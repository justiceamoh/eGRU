This notebook discusses how to implement a sequence-to-sequence deep recurrent network for an acoustic event detection task using eGRU cells as published in the paper ["An Optimized Recurrent Unit for Ultra-Low-Power Keyword Spotting"](https://arxiv.org/abs/1902.05026). The AED task under study here is cough detection.  

```python
%matplotlib inline

# Basic Imports
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

# PyTorch 
import torch as th
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

# Import Custom eGRU Modules
from eGRUModules import eGRUCell, QLinear

# Scikit-Learn Modules
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score
from sklearn.utils.class_weight import compute_class_weight
from scipy.signal import medfilt

# Audio Playback Modules
import librosa as lbr
from librosa.display import waveplot, specshow
from IPython.display import display,Image, Audio
from tqdm import tqdm, tqdm_notebook, trange
```

```python
th.manual_seed(7);
device = 'cuda' if th.cuda.is_available() else 'cpu'
```

```python
# Controls
use_quant   = True         # Use Quantization
use_binary  = True         # Convert to Binary
use_cweight = True         # Use Class Weights (for Imbalanced Classes)
```

```python
# Parameters
max_epoch  = 500
batch_size = 128
```

```python
# Dataset
maxlen  = 24
dataset = 'coughdataset.pkl'
```

```python
# Output Files
netname = 'net.pth'
logname = 'log.csv'
```

## Load Training Data

```python
# Get Data
def get_data():
    df   = pd.read_pickle(dataset)
    cmap = dict(enumerate(df.Class.cat.categories))
    print '' 
    print 'Entries in dataset: %d'%(len(df))
    print 'Class Map: ', cmap

    # Data Preparations
    # Truncation
    pad = lambda a, n: a[:,0: n] if a.shape[1] > n else np.hstack((a, np.zeros([a.shape[0],n - a.shape[1]])))
    Stft = df.Stft.apply(pad,args=(maxlen,))
            
    # Get X & Y Data
    x_data = np.dstack(Stft.values).transpose(2,1,0) # size: (batch_size, timesteps, dim)
    y_data = df.Class.cat.codes.values.astype('int32')

    return x_data,y_data,cmap
```

```python
# Get Event Selection 
def get_selection(s,thresh=30,bg=0,fg=1,fwin=9):
    ss = np.sum(s,axis=-1)        # Sum along dimension
    ss = ss/ss.max()              # Normalize
    yt = np.percentile(ss,thresh) # Get threshold as percentile of sums
    ys = bg*np.ones_like(ss)      # New array filled with background, bg 
    ys[ss>yt] = fg                # Highlight selections with foreground, fg
    ys = medfilt(ys,fwin)         # Median filter to smoothen with fwin window length
    return ys
```

```python
# Convert Training Data to PyTorch DataLoader
def get_dataloader(x_data,y_data,batch_size=32):
    # Convert to Torch Tensors
    x = th.from_numpy(x_data).float()
    y = th.from_numpy(y_data).long()

    # TensorDataset & Loader
    dataset = TensorDataset(x,y)
    loader  = DataLoader(dataset,batch_size=batch_size,shuffle=True,drop_last=True) 
    return loader
```

```python
x_data,y_data,cmap = get_data()

# Get Finer Sequential labels
if use_binary:
    y_data = np.array([get_selection(x,fg=y,bg=1) if y==0 else np.ones(24) for x,y in zip(x_data,y_data)])

# Binarize
if use_binary:
    y_data[y_data==2] = 1
    cmap = {0:'cough',1:'noise'}
    print 'New Map:', cmap

    
# Data Splitting Indices
indices = range(len(y_data))
itrain,itestn,_,_ = train_test_split(indices,y_data,test_size=0.3,random_state=32) 
itrain,ivalid,_,_ = train_test_split(itrain,y_data[itrain],test_size=0.25,random_state=32)

# Dataloaders
trainloader = get_dataloader(x_data[itrain],y_data[itrain],batch_size)
validloader = get_dataloader(x_data[ivalid],y_data[ivalid],batch_size)
```

## Network Training

```python
# Network Architecture
# Dimensions
steps,fbins = x_data.shape[1:]
nclass      = len(np.unique(y_data)) 
L1,L2,L3    = 32,20,16
```

```python
# Neural Network
class Network(nn.Module):
    def __init__(self,dev='cpu'):
        super(Network, self).__init__()
        self.dev  = dev
        self.gru1 = eGRUCell(fbins,L1,use_quant)
        self.gru2 = eGRUCell(L1,L2,use_quant)            
        self.fc3  = QLinear(L2,L3)      
        self.fc4  = QLinear(L3,nclass)

    def forward(self,inputs):
        blen = inputs.size(0)
        tlen = inputs.size(1)
        
        h1 = th.Tensor(blen,L1).uniform_(-1.0,1.0).to(self.dev)
        h2 = th.Tensor(blen,L2).uniform_(-1.0,1.0).to(self.dev)
        h  = []
        
        for t in range(tlen):
            x  = inputs[:,t,:]
            h1 = self.gru1(x,h1)
            h2 = self.gru2(h1,h2)
            h3 = self.fc3(h2).clamp(min=0)
            h4 = self.fc4(h3)
            h.append(h4)
        y = th.stack(h).permute(1,0,2)
        return F.log_softmax(y,dim=-1)
```

## Train Network

```python
# Instantiate
net = Network(dev=device)
opt = optim.Adam(net.parameters(),lr=0.001)
```

```python
# Sequential Cross-entropy
def crt(x,y):
    xp = x.permute(0,2,1)
    return F.nll_loss(xp,y,weight=cweight)

# Mode Prediction
def get_pred(outs):
    pred   = th.argmax(outs,-1)
    return pred
```

```python
# Training Loop Function
LOSS,ACCU = [],[]
def do_training(max_epoch):
    for epoch in tqdm_notebook(range(max_epoch)):
        rloss = []
        acc   = 0.0
        for i,data in enumerate(trainloader):
            x,y = data
            opt.zero_grad()

            outs = net(x)
            loss = crt(outs,y)
            loss.backward()
            opt.step()
            
            rloss.append(loss.item())
         
        total,correct = 0.,0.
        with th.no_grad():
            for data in validloader:
                x,y    = data
                outs   = net(x)
                pred   = get_pred(outs)
                total += y.flatten().size(0)
                correct+= (pred.flatten() == y.flatten()).sum().item()        
        
        LOSS.append(np.mean(rloss))
        ACCU.append(correct/total)
        tqdm.write('Epoch %d/%d, Loss: %.4f, Accu: %.4f'%(epoch+1,max_epoch,LOSS[-1],ACCU[-1])) 
    return
```

```python
# Run Training
do_training(max_epoch)
```

```python
# Training Curves
plt.figure(figsize=(8,3))
plt.subplot(121)
plt.plot(range(len(LOSS)),LOSS,'g'); plt.title('Loss');
plt.subplot(122)
plt.plot(range(len(ACCU)),ACCU,); plt.title('Accuracy');
```

## Evaluation - Concatenated Input Test

```python
# Sampling Parameters
Fs   = 8000
nfft = 128

# Numpy Stft Helper Functions
fft  = lambda x: np.abs(np.fft.fft(x))[:,:x.shape[1]/2].T
stft = lambda x,n,s: fft(np.array(zip(*[x[i::s] for i in range(n)])))

def get_onehot(y):
    yhot = np.zeros((y.size, y.max()+1))
    yhot[np.arange(y.size),y] = 1
    return yhot
```

```python
# Get Continuous Test Audio
def get_audio(K=None,thresh=5,nidx=2):
    df = pd.read_pickle(dataset)
    waves = df.Wave
    label = df.Class.cat.codes.values.astype('int32')
    if use_binary:
        label[label==2] = 1
        
    if K is None:
        w_data = waves[itestn].tolist()
        l_data = label[itestn].tolist()
    else:
        Kidx   = np.random.choice(len(itestn),K)
        itdx   = np.array(itestn)
        w_data = waves[itdx[Kidx]].tolist()
        l_data = label[itdx[Kidx]].tolist()
    
    l_data = [l*np.ones(len(w)) for l,w in zip(l_data,w_data)]
    
    # Generated Noise for Padding
    sigma = np.percentile(np.hstack(np.abs(w_data)),thresh)
    wlens = map(len,w_data)                               # lengths of test examples               
    nlens = np.random.randint(min(wlens),max(wlens),K)  # random lengths of padding noise
    noise = [sigma*np.random.randn(l) for l in nlens]
    nlabl = [nidx*np.ones(l) for l in nlens]
    
    # Interleave
    stream       = w_data + noise
    stream[::2]  = w_data
    stream[1::2] = noise
    
    labels       = l_data + nlabl
    labels[::2]  = l_data
    labels[1::2] = nlabl

    # Concatenate
    stream = np.concatenate(stream)
    labels = np.concatenate(labels).astype('int32')
    return stream,labels
```

```python
# Get Audio & Spectrogram
w_test,y_test = get_audio(K=20,nidx=1)
y_test = get_onehot(y_test)
x_test = stft(w_test,nfft,nfft-4)
t_test = lbr.core.samples_to_time(range(len(w_test)),Fs)
```

```python
# Plot Audio & Spectrogram
plt.figure(figsize=(15,4))
plt.subplot(211)
specshow(x_test,cmap='viridis')
plt.subplot(212)
waveplot(w_test,sr=Fs);
plt.subplots_adjust(hspace=0)

display(Audio(data=w_test,rate=Fs))
```

```python
# Pass through network
x_test = th.tensor(x_test.T[None,:],dtype=th.float)

with th.no_grad():
    y_pred = th.softmax(net(x_test),-1)
    
y_pred = y_pred.data.numpy().squeeze()
t_pred = lbr.core.frames_to_time(np.arange(len(y_pred)),Fs,nfft)
```

```python
# Plot Waveform Predictions
plt.figure(figsize=(15,4))
plt.subplot(211)
waveplot(w_test,Fs);
plt.plot(t_test,0.5*y_test[:,0])
plt.legend(['cough','audio'],loc=1)

plt.subplot(212)
plt.margins(x=0)
plt.plot(t_pred,y_pred[:,1])
plt.plot(t_pred,y_pred[:,0])
plt.legend(['noise','cough'],loc=1)
plt.subplots_adjust(hspace=0)
```

## Evaluation - Streaming Input Test
Here, we evaluated the network using an external audio recording from subjects outside of the training dataset.

```python
# Data Files
wavfile   = 'coughrecording.wav'
labfile   = 'coughrecording.txt'
```

```python
# Load & Extract Recording Features
# Get Data & Features
x,_ = lbr.load(wavfile,Fs)           # Load Data
xs  = stft(x,nfft,nfft-4)  # Get Features

# Get Time
t   = lbr.core.samples_to_time(np.arange(len(x)),Fs)
ts  = lbr.core.frames_to_time(np.arange(len(xs.T)),Fs,nfft)
```

```python
# Get Labels
dl    = pd.read_csv(labfile,delimiter='\t',names=['Start','Stop','Label'])
lo,hi = dl.Start.values,dl.Stop.values

# Specify Categories
dl['Label'] = dl['Label'].astype('category')

# Get Label Vector
l     = dl.Label.cat.codes.values.astype('int32')
```

```python
# Build Ranges for Look Up
lranges = [zip(lo[l==k],hi[l==k]) for k in np.unique(l)]  # Get list of tuple of ranges for each label

# Look Up label by range
def rlookup(t):
    for i,r in enumerate(lranges):
        if any(lo <= t <= hi for (lo,hi) in r):
            return i
    return i+1
```

```python
# Get Label Vector
y = np.array([rlookup(i) for i in t])
yhot = get_onehot(y) # One Hot Encoding
```

```python
# Simulate Streaming Input
xt = xs.T
xt = th.Tensor(xt[None,:])
with th.no_grad():
    odata = th.softmax(net(xt),-1)

hdata = odata.data.numpy().squeeze()
```

```python
# Plot Labels
plt.figure(figsize=(12,4))
plt.subplot(211)
waveplot(x,Fs); plt.ylim([-0.5,0.5]);
plt.plot(t,0.4*yhot[:,:-1],alpha=0.8);
plt.legend(['cough','audio'],loc=2);
plt.margins(0)
plt.gca().get_xaxis().set_visible(False)

plt.subplot(212);
plt.plot(ts,hdata[:,-1])
plt.plot(ts,hdata[:,:-1])
plt.margins(0),plt.ylim([-0.02,1.02])
plt.tight_layout()
plt.legend(['noise','cough'],loc=2);
plt.subplots_adjust(hspace=0)

display(Audio(data=x,rate=Fs))
```
